```python
df = pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
columns = ["City","Region","Country","AirQuality","WaterPollution"]
df.columns = columns
df.head()
```

```python
df['Region'] = df.Region.map(lambda x: x.replace('"',""))
df.head()
```

```python
df['Country'] = df.Country.map(lambda x: x.replace('"',""))
df.head()
```

```python
df.to_csv('updated_world_cities_air_water_quality.csv',index=False)
```

```python
df['Country'] = df.Country.map(lambda x: x.strip())
df.head()
```

```python
df[df.Country == 'Germany']
```

```python
df1 = pd.pivot_table(df,index=['Country','City'],values = ['AirQuality', 'WaterPollution'],aggfunc='mean')
df1
```

```python
df1.index
```

```python
df1.loc['Germany']
```

```python
df2 = df1.copy()
df2.head()
```

```python
df2['AirPollution'] = 100-df2['AirQuality'] # AirPollution ustuniga teskari ko'rsatgichli ustun yaratish
df2.head()
```

```python
df2['WaterQuality'] = 100-df2['WaterPollution'] # WaterQuality ustuniga teskari ko'rsatgichli ustun yaratish
df2
```

```python
df2.head()
```

```python
df3 = pd.DataFrame(df2.values, index = df2.index, columns= [['AIR', 'AIR', 'WATER', 'WATER'], ['AirQuality', 'AirPollution', 'WaterQuality', 'WaterPollution']]) #
df3
```
